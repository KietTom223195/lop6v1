**Môn học:** Trí tuệ nhân tạo / Học máy

## 1. Mục tiêu bài tập
- Hiểu và áp dụng được thư viện Keras / TensorFlow để xây dựng một mạng nơ-ron nhân tạo (ANN) cơ bản.
- Biết cách chuẩn bị dữ liệu phân loại nhiều lớp (Multi-class Classification) bằng phương pháp mã hóa One-Hot Encoding.
- Làm quen với các khái niệm cốt lõi trong huấn luyện mô hình: Tầng ẩn (Hidden Layer), Hàm kích hoạt (Activation Function), Hàm mất mát (Loss function), Chu kỳ huấn luyện (Epochs) và Đánh giá mô hình (Evaluation).

## 2. Import các thư viện cần thiết

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

## BƯỚC 1: TẢI VÀ CHUẨN BỊ DỮ LIỆU

```python
iris = load_iris()
X = iris.data  
y = iris.target.reshape(-1, 1)  

# Mã hóa nhãn sang dạng One-Hot Vector
encoder = OneHotEncoder(sparse_output=False)
y_onehot = encoder.fit_transform(y)

# Chia dữ liệu thành tập Train (80%) và Test (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y_onehot, test_size=0.2, random_state=42)

print(f"Kích thước tập Train: {X_train.shape}")
print(f"Kích thước tập Test: {X_test.shape}")
```

## BƯỚC 2: XÂY DỰNG CẤU TRÚC MẠNG ANN
Mô hình dạng chuỗi (Sequential) gồm:
- Tầng ẩn số 1: 8 nơ-ron, hàm kích hoạt relu, nhận đầu vào là 4 đặc trưng.
- Tầng ẩn số 2: 6 nơ-ron, hàm kích hoạt relu.
- Tầng đầu ra: 3 nơ-ron (phân loại 3 lớp), hàm kích hoạt softmax.

```python
model = Sequential()
model.add(Dense(8, input_shape=(4,), activation='relu')) # Tầng ẩn 1
model.add(Dense(6, activation='relu'))                   # Tầng ẩn 2
model.add(Dense(3, activation='softmax'))                # Tầng đầu ra

model.summary()
```

## BƯỚC 3 & 4: CẤU HÌNH QUY TRÌNH HUẤN LUYỆN & HUẤN LUYỆN MÔ HÌNH
- Thuật toán tối ưu: Adam (learning rate = 0.01)
- Hàm mất mát: categorical_crossentropy
- Số chu kỳ huấn luyện epochs = 50
- Kích thước gói batch_size = 10

```python
model.compile(optimizer=Adam(learning_rate=0.01),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print("\n--- Bắt đầu huấn luyện mạng ANN ---")
history = model.fit(X_train, y_train, epochs=50, batch_size=10, verbose=1)
```

## BƯỚC 5: ĐÁNH GIÁ MÔ HÌNH TRÊN TẬP KIỂM TRA (TEST SET)

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"\nĐỘ CHÍNH XÁC TRÊN TẬP KIỂM TRA: {accuracy * 100:.2f}%")
```

## BƯỚC 6: DỰ ĐOÁN THỰC TẾ TRÊN MẪU HOA MỚI
- Chiều dài lá đài: 5.1cm
- Chiều rộng lá đài: 3.5cm
- Chiều dài cánh hoa: 1.4cm
- Chiều rộng cánh hoa: 0.2cm

```python
hoa_moi = np.array([[5.1, 3.5, 1.4, 0.2]])
du_doan = model.predict(hoa_moi)
loai_hoa_du_doan = np.argmax(du_doan)

print(f"Xác suất dự đoán cho 3 loài hoa lần lượt là: {du_doan}")
print(f"-> Kết quả dự đoán loài hoa mang nhãn số: {loai_hoa_du_doan} ({iris.target_names[loai_hoa_du_doan].upper()})")
```
